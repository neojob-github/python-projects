Найдем, какие сорта вин лидирует по рейтингу:

```python
uniq_variety = df.variety.unique()
top_wine_by_raiting = {}
for variety in uniq_variety:
    raiting = df[df.variety == variety]["points"].mean()
    top_wine_by_raiting.update({variety : raiting})
top_wine_by_raiting = sorted(top_wine_by_raiting.items(), key=lambda x: x[1])
top_wine_by_raiting
```

Из этого можно сделать вывод, что наибольший рейтинг вина выходит из винограда сорта <span style="color:#00ffff">"Tokay"

Узнаем, какие сорта лидируют по регионам:

```python
def get_regions(dataframe : pd.DataFrame, region_name : str):
    if region_name =="region_1":
        regions_series = pd.Series({"regions" : list(dataframe.region_1.unique())})
        regions_series.drop_duplicates(inplace=True)
        regions = list(regions_series)[0]
    else:
        regions_series = pd.Series({"regions" : list(dataframe.region_2.unique())})
        regions_series.drop_duplicates(inplace=True)
        regions = list(regions_series)[0]
    return regions


regions_1_raiting = {}
regions_2_raiting = {}
all_region_raitings = {}
new_df = pd.DataFrame()

regions = get_regions(df, region_name="region_1")
for region in regions:
    regions_1_raiting.update({region : df[df.region_1 == region]["points"].max()})

regions = get_regions(df, region_name="region_2")
for region in regions:
    regions_2_raiting.update({region : df[df.region_2 == region]["points"].max()})

for key, value in regions_2_raiting.items():
    if key in regions_1_raiting:
        if regions_1_raiting.get(key) < value:
            all_region_raitings.update({key : value})

for key, value in regions_1_raiting.items():
    if key not in all_region_raitings:
        all_region_raitings.update({key : value})



for key, value in all_region_raitings.items():
    new_df = df[df.region_1 == key][df.points == value][["region_1", "points", "variety"]].drop_duplicates()
    new_df = new_df.reset_index()
    for index, row in new_df.iterrows():
        print(f"В регионе {row['region_1']} вино сорта {row['variety']} имеет лучший рейтинг равный {row['points']}")


```

Бюджетный сегмент вин можно определить как всё вино цена которого ниже среднерыночной стоимости:

```python
def draw_circle_diagramm(budget_wines_count, premium_wines_count):
    labels = ["Budget", "Premium"]
    values = [budget_wines_count, premium_wines_count]
    colors = ["#20DE71", "#FDC3E0"]
    explode = [0, 0.07]
    plt.title("Wine segments", fontsize=20)
    plt.pie(values,labels=labels,colors=colors,explode=explode,startangle=80, shadow=True, autopct='%1.2f%%', textprops={'fontsize': 16})
    plt.axis("equal")
    plt.show()

mean_wine_price = df["price"].mean()
budget_wines_count = df[df.price < mean_wine_price].count()[0]
premium_wines_count = df.count()[0] - budget_wines_count
draw_circle_diagramm(budget_wines_count, premium_wines_count)
print(f"Всего бюджетного вина: {budget_wines_count}шт.\nВсего премиум вина: {premium_wines_count}шт.\nВсего вина: {budget_wines_count + premium_wines_count}шт.")
```

Из приведённой выше информации можно сделать вывод, что на исследуемом рынке преобладает вино бюджетного класса.<br>
На него приходится <span style="color:#20DE71">66.86%</span> от всего вина.<br>
Остальные <span style="color:#FDC3E0">33.14%</span> делят между собой вина класса "премиум" и "люкс".

```python
plt.figure (figsize = (10, 10))
region_t1 = list(df["region_1"].unique())
region_t2 = list(df["region_2"].unique())
regions = []
for region in region_t2:
    if region not in region_t1:
        regions.append(region)
regions = regions + region_t1
regions.sort()
regions
k = 0
for region in regions:
    prices = df[df.region_1 == region]["price"]
    mean_price = prices.mean()
    if type(mean_price) != np.float_:
        mean_price = df[df.region_2 == region]["price"].mean()
    plt.bar(x = k, y = mean_price, height=5, width=5, color="blue")
    k += 1
plt.title("Распределение средней цены на вино в регионах", fontsize=20)
plt.xlabel("Выборка вина")
plt.ylabel("Средняя цена")
plt.show()
plt.figure (figsize = (10, 10))
k = 0
for region in regions:
    prices = df[df.region_1 == region]["price"]
    mean_price = prices.mean()
    std_price = prices.std()
    if type(mean_price) != np.float_:
        mean_price = df[df.region_2 == region]["price"].mean()
        std_price = prices.std()

    plt.bar(x = k, y = std_price, height=5, width=5, color="red")
    k += 1
plt.title("Распределение std на цены на вино в регионах", fontsize=20)
plt.xlabel("Выборка вина")
plt.ylabel("Отклонение")
plt.show()
plt.figure (figsize = (10, 10))
k = 0
for region in regions:
    prices = df[df.region_1 == region]["price"]
    mean_price = prices.mean()
    var_price = prices.var()
    if type(mean_price) != np.float_:
        mean_price = df[df.region_2 == region]["price"].mean()
        var_price = prices.var()
    if var_price > 100000:
        var_price = 0
    plt.bar(x = k, y = var_price, height=285.714285714, width=5, color="green")
    k += 1
plt.title("Распределение дисперсии на цены на вино в регионах", fontsize=20)
plt.xlabel("Выборка вина")
plt.ylabel("Дисперсия")
plt.show()
```

1. Высокая средняя цена показывает то, как распределяются цены в зависимости от региона в котором продают вино.
2. Высокие значения std (стандратного отклонения) показывают, что точки данных могут распространяться далеко от среднего значения, а небольшое стандартное отклонение указывает на то, что они тесно сгруппированы вокруг среднего значения.
3. Высокая дисперсия указывает на то, что разброс данных очень велик и значения находятся далеко от среднего значения.

#  <a name="5">5. Проверка гипотез </a>

* H0: Средние пользовательские рейтинги красного и белого вина одинаковые.
* H1: Средние пользовательские рейтинги красного и белого вина разные.


* H0: Средние цены двух популярных сортов вина одинаковые.
* H1: Средние цены двух популярных сортов вина разные.


Нормальное распределение обычно связано с правилом трех сигм.

68% данных находятся в пределах 1 стандартного отклонения (σ) от среднего значения (μ)<br>
95% данных находятся в пределах 2 стандартных отклонений (σ) от среднего значения (μ)<br>
99,7% данных находятся в пределах 3 стандартных отклонений (σ) от среднего значения (μ)<br>

```python
mean_red_wine_points = df[df.color == "red"]["points"].mean()
mean_white_wine_points = df[df.color == "white"]["points"].mean()
alpha = 0.05
print(f"Средние оценки красного вина равны: {mean_red_wine_points}")
print(f"Средние оценки белого вина равны: {mean_white_wine_points}")
answer = ((mean_red_wine_points ** 2) / (mean_white_wine_points ** 2)) - 1
print("H0 is True!") if answer > 0.05 else print("H0 is False!")

```

Данный эксперимент отвергает гипотезу H0, т.к. в сравнении вино не переходит пороговые значения p-value

```python
mean_pinot_price = df[df.variety == "Pinot Noir"]["price"].mean()
mean_cabernet_price = df[df.variety == "Cabernet Sauvignon"]["price"].mean()
print(f"Средние оценки Pinot Noir равны: {mean_pinot_price}")
print(f"Средние оценки Cabernet Sauvignon равны: {mean_cabernet_price}")
answer = ((mean_pinot_price ** 2) / (mean_cabernet_price ** 2)) - 1
print("H0 is True!") if answer > 0.05 else print("H0 is False!")
```

Данный эксперимент подтверждает гипотезу H0, т.к. в сравнении вино переходит пороговые значения p-value
